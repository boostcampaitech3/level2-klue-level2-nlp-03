```python
import pickle
def num_to_label(n):
    with open('/opt/ml/tests/level2-klue-level2-nlp-03/custom_baseline_ver3/code/dict_num_to_label.pkl', 'rb') as f:
        dict_num_to_label = pickle.load(f)
    origin_label = dict_num_to_label[n]
    return origin_label
import torch.nn.functional as F
import torch
first_ens = pd.read_csv('./results/first.csv')
first_ens['probs'] = first_ens['probs'].apply(lambda x : to_nparray(x)*0.5) + second['probs'].apply(lambda x : to_nparray(x)*0.5)
first_ens['softmax'] = first_ens['probs'].apply(lambda x : F.softmax(torch.tensor(x), dim=-1).detach().cpu().numpy())
first_ens['pred_label'] = first_ens['softmax'].apply(lambda x : num_to_label(np.argmax(x)))
```

```python
first_sec_ens = pd.read_csv('./results/first.csv')
first_sec_ens['probs'] = first_sec_ens['probs'].apply(lambda x : to_nparray(x)*0.5) + third['probs'].apply(lambda x : to_nparray(x)*0.5)
first_sec_ens['softmax'] = first_sec_ens['probs'].apply(lambda x : F.softmax(torch.tensor(x), dim=-1).detach().cpu().numpy())
first_sec_ens['pred_label'] = first_sec_ens['softmax'].apply(lambda x : num_to_label(np.argmax(x)))
```
